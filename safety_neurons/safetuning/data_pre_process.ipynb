```python
with open('./Llama_attack_filtered.json', 'r') as f:
    data = json.load(f)
```

```python
result = {}

for attack in data:
    r = []
    for item in data[attack]:
        for key in item:
            if key!='prompt':
                response = item[key]
        r.append({'prompt': item['prompt'], 'response': response})
    result[attack] = r
                
```
